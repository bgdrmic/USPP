```python
import pandas as pd
import re
import os
```

```python
datasets = ['index_articles.csv', 'index_articles2.csv', 'index_articles3.csv', 'index_articles4.csv', 'index_articles5.csv', 'index_articles6.csv']
```

```python
df = None
for i, ds in enumerate(datasets):
    if i == 0:
        df = pd.read_csv(ds)
        continue
    df2 = pd.read_csv(ds)
    df = pd.concat([df, df2], axis=0)
df.to_csv('combined_articles.csv', index=False)
```
